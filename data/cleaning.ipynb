```python
import pandas as pd
import os
import json
import requests

#Cleaning JSON File
with open("raw/yelp_raw.json") as f:
    all_businesses = json.load(f) #90969 restaurants

del_columns = [ 
    "id", "image_url", "is_closed", "url", "coordinates", "location",
    "phone", "display_phone", "distance", "business_hours", "attributes"]

for business in all_businesses: #deleting unnecessary columns
    for col in del_columns:
        business.pop(col, None)

for business in all_businesses: #cleaning data
    alias_list = []
    title_list = []
    for category in business["categories"]:
        alias_list.append(category["alias"])
        title_list.append(category["title"])
    
    business["categories_alias"] = alias_list
    business["categories"] = title_list

    business["name_alias"] = business.get("alias")
    business.pop("alias", None)

    location = business["city_query"].split(", ")
    business["city"] = location[0]
    business["state"] = location[1]



df = pd.json_normalize(all_businesses) #df formation
df = df[["name", "rating", "price", "categories", "city", "state", "review_count", "transactions", "name_alias", "categories_alias"]]

df.to_csv("yelp_cleaned.csv")


```
